```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Summer 2024 CS Class Schedule Data from this URL
url = 'https://web.cs.kent.edu/schedule/index.php'
# Consider the read_html() method, but remember multiple tables are returned and the interesting data might not be in the first table
nfd = pd.read_html(url) 
print(len(nfd)) # prints length of  not filtered data
df = nfd[1] #because there are two data frames converted we have to put the one we want into df.
```
